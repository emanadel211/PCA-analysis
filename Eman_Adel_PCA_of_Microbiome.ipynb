```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
link1 = "https://raw.githubusercontent.com/ahmedmoustafa/datasets/main/healthy_adults_microbiome/healthy_adults_microbiome.tsv"
#path = "relative_abundance.tsv"
relative_abundance =  pd.read_csv(link1, sep="\t", index_col = 0)
relative_abundance.shape
```

```python
relative_abundance.head()
```

```python
relativeabundance = relative_abundance.transpose()
relativeabundance.shape
```

```python
relativeabundance.head()
```

```python
sample = relativeabundance.index
```

```python
features = relativeabundance.values
```

```python
features
```

Using a scree plot and variance ratio values for the first 15 components to determine the number of components to use in the PCA analysis

```python
pca15 = PCA(n_components = 15)
principal_components15 = pca15.fit_transform(features)
```

```python
import numpy as np
```

```python
#percentage of the expalined varince ration of the first 15 components
principal_variation15 = np.round(pca15.explained_variance_ratio_*100, decimals = 1)
print(principal_variation15)
```

```python
sum(principal_variation15)
```

```python
# scree plot of the first 15 components
labels = ['PC'+ str(x) for x in range (1, len(principal_variation15) + 1)]
plt.bar(x=range(1, len(principal_variation15)+1), height = principal_variation15)
plt.ylabel("Percentage of Explained Variance")
plt.xlabel("Principal Component")
plt.title("Scree Plot")
plt.xticks(range(1, len(principal_variation15) + 1), labels)
plt.show
```

The scree plot shows that the percentage of data explained by the 15 pricnipal components is quite low accounting for around 68% only, almost 50% of which is explained by the first 3 prinicpal components, and the plot started to plateau from PC5, so for the sake of easier interpretation and as I will not gain much insight if I increased the number of prinicipal components, I will use 3 principal components for the analysis

```python
# performing 3PCs PCA analysis
pca = PCA(n_components=3)
principal_components = pca.fit_transform(features)
principal_components
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2','PC3'])
principal_df['sample_id'] = sample.values
principal_df.head()
```

```python
principal_df.shape
```

```python
link2 = "https://raw.githubusercontent.com/ahmedmoustafa/datasets/main/healthy_adults_microbiome/healthy_adults_metadata.tsv"
metadata = pd.read_csv(link2, sep = '\t', index_col = 0)
metadata.shape
```

```python
metadata.head()
```

```python
metadata.info() # columns containing null-value are the ones that have non-value < 10264
```

```python
# removing the age and BMI column
metadata_age = metadata.drop(["age", "BMI"], axis = 1)
```

```python
metadata_age.shape
```

```python
#cleaning up missed data -  deleting rows that contain Nan values
clean_metadata = metadata_age.dropna(axis=0)
```

```python
clean_metadata.shape
```

```python
#merging on the sample identifiers
aligned_data = pd.merge(principal_df, clean_metadata, how = 'inner', on = 'sample_id')
```

```python
aligned_data.head()
```

```python
aligned_data.shape
```

# Loading investigation

```python
labels = relative_abundance.index.to_series()
```

```python
loading = pca.components_.T  #the columns are the loading scores for every PC and the rows are the variables or species
```

```python
loading_scores = pd.DataFrame (data = loading, columns = ['PC1', 'PC2', 'PC3'])
loading_scores["species"] = labels.values
loading_scores.head()
```

```python
#getting the top ten species for PC1
loading_pc1 = pd.Series(pca.components_[0], index = labels)
loading_pc1
```

```python
sorted_loading_pc1 = loading_pc1.abs().sort_values(ascending = False)
sorted_loading_pc1
```

```python
print(sorted_loading_pc1[:10]) #absolute loading scores
```

The first species is the one that have the highest contribution to PC1

```python
#getting the top ten species for PC1
top_10_species = sorted_loading_pc1[0:10].index.values
print(loading_pc1[top_10_species]) #real loading scores
```

The first species is the one that have the highest contribution to PC2

```python
#getting the top ten species for PC2
loading_pc2 = pd.Series(pca.components_[1], index = labels)
sorted_loading_pc2 = loading_pc2.abs().sort_values(ascending = False)
top_10_species = sorted_loading_pc2[0:10].index.values
print(loading_pc2[top_10_species]) #print the species name and the coresponding loading scores
```

```python
# visualizing the top ten species to PC1 using the absolute values
plt.figure(figsize = (10,5))
plt.bar(x= sorted_loading_pc1.index[:10], height = sorted_loading_pc1.values[:10], color ='maroon')
plt.xlabel("Species")
plt.ylabel("Loading Score")
plt.title("PC1 Loading Scores")
plt.xticks(rotation=90, ha="right")
plt.show()
```

```python
# visualizing the top ten species to PC2 using the absolute values
plt.figure(figsize = (10,5))
plt.bar(x= sorted_loading_pc2.index[:10], height = sorted_loading_pc2.values[:10], color ='green')
plt.xlabel("Species")
plt.ylabel("Loading Score")
plt.title("PC2 Loading Scores")
plt.xticks(rotation=90, ha="right")
plt.show()
```

# Regression Analysis

```python
import statsmodels.api as sm
```

```python
aligned_data.head()
```

```python
#removing sample_id and subject_id columns for easier interpretation
regression_data = aligned_data.drop(['sample_id','subject_id'], axis = 1)
```

```python
regression_data.head()
```

```python
#converting categorical data into numerical data
df_encoded = pd.get_dummies(regression_data, drop_first=True)
df_encoded.shape
```

```python
df_encoded.head()
```

```python
df_encoded.columns
```

```python
#dependenat variable is sex and independent variable are PCs
y_pc1 = regression_data['PC1']
x_sex = sm.add_constant(df_encoded['gender_male'])
model_gender = sm.OLS(y_pc1, x_sex).fit()
```

```python
print(model_gender.summary())
```

The constant p-value is significant, but the male p-value is not significant, indicating that the female gender may have an effect on the microbiome variation not the male (the female p-value needs to be calculated)

```python
#including only 7 countries
x_country = sm.add_constant(df_encoded[['country_BRN', 'country_CAN', 'country_CHN', 'country_DNK', 'country_ESP', 'country_ITA', 'country_USA']])
model_gender = sm.OLS(y_pc1, x_country).fit()
```

```python
print(model_gender.summary())
```

- The constant p-value is significant, but not all countries have a significant p-value, indicating that some countries have an infleunce on the microbiome association
- The slope value is small, indicating a small impact of the country on the microbiome variation
