```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
link1 = "https://raw.githubusercontent.com/ahmedmoustafa/datasets/main/iHMP_IBDMDB_2019/species.tsv"
#path = "relative_abundance.tsv"
relative_abundance = pd.read_csv(link1, sep="\t", index_col = 0)
relative_abundance.shape
```

```python
relative_abundance.head()
```

```python
sample = relative_abundance.index
```

```python
relative_abundance.head()
```

```python
features = relative_abundance.values
```

```python
features
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(features)
principal_components
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.score
```

```python
pca.components_.T
```

```python
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
principal_df['Sample'] = sample.values
principal_df.head()
```

```python
link2 = "https://raw.githubusercontent.com/ahmedmoustafa/datasets/main/iHMP_IBDMDB_2019/metadata.tsv"
# path = "metadata.tsv"
metadata = pd.read_csv(link2, sep="\t", index_col = 0)
```

```python
metadata.shape
```

```python
metadata.head()
```

```python
metadata.info()
```

columns that contain null values are those that have non-null values < 382, including: consnet age, Age at diagnosis, fecalcal, BMI at baseline, height at baseline, weight at baseline, smoking status.

```python
#cleaning up data
metadata = metadata.drop(["smoking_status", "age_at_diagnosis", "BMI_at_baseline", "Height_at_baseline", "Weight_at_baseline", "consent_age" ], axis = 1)
```

```python
metadata.shape
```

```python
#cleaning up missed data -  deleting rows that contain Nan values
clean_metadata = metadata.dropna(axis=0)
```

```python
clean_metadata.shape
```

```python
clean_metadata.head()
```

```python
clean_metadata.columns
```

```python
relative_abundance.columns
```

```python
#merging on the sample identifiers
aligned_data = pd.merge(principal_df, clean_metadata, how = 'inner', on = 'Sample')
```

```python
aligned_data.head()
```

```python
aligned_data.shape
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data = aligned_data, x = 'PC1', y = 'PC2', hue = 'Study.Group')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title("PCA for microbiome for IBS")
plt.show()
```

As shown in the scatter plot there is no clustering in the data, so further analysis is not meaningful
